```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from collections import Counter

# 1. Generate imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0, random_state=1)

# Create a DataFrame for visualization
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
df['Class'] = y

# 2. Visualize class distribution before SMOTE
sns.countplot(x='Class', data=df)
plt.title('Class Distribution Before SMOTE')
plt.show()

# 3. Apply SMOTE to balance the dataset
oversample = SMOTE()
X_smote, y_smote = oversample.fit_resample(X, y)

# Create a DataFrame for the balanced data
df_smote = pd.DataFrame(X_smote, columns=['Feature 1', 'Feature 2'])
df_smote['Class'] = y_smote

# 4. Visualize class distribution after SMOTE
sns.countplot(x='Class', data=df_smote)
plt.title('Class Distribution After SMOTE')
plt.show()

# 5. Print class distribution before and after SMOTE
print("Original dataset shape %s" % Counter(y))
print("Resampled dataset shape %s" % Counter(y_smote))
```
